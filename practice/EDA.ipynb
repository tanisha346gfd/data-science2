FILLING THE MISSING VALUES- IMPUTATION


```python
df3.head()
```

```python
mean_value=df3["reviews_per_month"].mean()
```

```python
df3["reviews_per_month"].fillna(mean_value,inplace=True)
```

```python
df3.nunique()
```

```python
df3.isna().sum()
```

```python
df3["last_review"].fillna(df3["last_review"].value_counts().index[0],inplace=True)
```

```python
df3.isnull().sum()
```
